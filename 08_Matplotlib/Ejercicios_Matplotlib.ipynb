Representa la función $\frac{D}{(x^2+1)}$ , donde D es el día de tu cumpleaños  
Hazque el título sea tenga una fuente: serif de color blue y tamaño 16

<img src="f1.png" ALIGN="left">


```python
import matplotlib.pyplot as plt
import numpy as np
font = {'family':'serif','color':'blue','size':15}

x = np.arange(-4,4, 0.1) 
y = 25 / (x * x+ 1)
plt.title("Carlos Hernández", loc = 'center', fontdict = font)
plt.plot(x,y,color='g',label=r'$\frac{25}{{(x^2+1)}}$')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.xticks(range(-4, 5))
plt.grid()
plt.show()
```

<b><font color='red'>Ejercicio 05</font></b>  

Haz la gráfica del analisis de frecuencias realizado del archivo "../04_RW/el_quijote.txt"  
tienes que obtener la siguiente gráfica:  
<img src="f6.png" ALIGN="left">

```python
abc="ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
archivo="el_quijote.txt"
letter_counts = {}
my_array = []
char_array = [char for char in abc]

with open(archivo, 'r', encoding='utf-8') as file:
    text = file.read().upper()
countTotal = 0

for letra in abc:
    letter_counts[letra] = 0

for letter in text:
    if letter.upper() in abc:
        countTotal+= 1;
        letter_counts[letter] += 1

for letter, count in letter_counts.items():
    my_array.append(count/countTotal)
    
plt.bar(char_array,my_array)
plt.plot()
```
