Crea un archivo llamado calendario.xlsx. Haz que tenga 12 hojas, llamadas
["enero","febrero","marzo","abril","mayo","junio","julio","agosto","septiembre","octubre","noviembre","diciembre"]
con los siguientes días por hoja
[31,28,31,30,31,30,31,31,30,31,30,31]
haz que apareza un calendario de ese mes en cada hoja como creas conveniente, es decir:

    L M X J V S D
    1 2 3 4 5 6 7
    ....
    
    o
    
    1 2 3 4 5 6 ....31
    
    o
    
    1
    2
    3
    4
    .
    .
    

```python
import openpyxl

wb = openpyxl.Workbook()

default_sheet = wb.active
wb.remove(default_sheet)

meses = ['Enero', 'Febrero', 'Marzo', 'Abril','Mayo', 'Junio', 'Julio', 'Agosto','Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

feb = False
for mes in meses:
    ws = wb.create_sheet(title=mes)
    dias_semana = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
    for col_num, dia in enumerate(dias_semana, start=1):
        ws.cell(row=1, column=col_num, value=dia)
    
    if meses.index(mes) % 2 != 0:
        if feb == False:
            numeros = list(range(1, 29))
            feb = True
        else:
            numeros = list(range(1, 31))
    else:
        numeros = list(range(1, 32)) 
    
    filaInicial = 2 
    columnaInicial = 1  
    for numero in numeros:
        ws.cell(row=filaInicial, column=columnaInicial, value=numero)
        columnaInicial += 1  
        if columnaInicial > 7:  
            columnaInicial = 1
            filaInicial += 1 
wb.save('calendario.xlsx')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import yfinance as yf
```

```python
#Si queremos los calores de CaixaBanck
start='2021-01-01'
end='2021-06-01'
df =  yf.download('CABK',start,end)
df.head()
```

Crea una nueva columna llamada que muestre el valor medio, es decir (Close+Open)/2 y haz una gráfica que muestre los tres valores:

```python
"""Esribe tu código aquí"""
df['medio']=(df['Close']+df['Open']) / 2
df.head()
```

Utiliza, hlines, fill_between de la libreria matplotlib para obtener el siguiente resultado:

```python
"""Esribe tu código aquí"""
open_prices = df['Open']
low_prices = df['Low']
close_prices = df['Close']

plt.figure(figsize=(10, 6))
plt.plot(df.index, open_prices, label='Open')
plt.plot(df.index, low_prices, label='Low')
plt.plot(df.index, close_prices, label='Close')
plt.show()
```

<img src="PAND2.png">  

crea una nueva gráfica en la que se muestre la diferencia entre la apertura y el cierre, así como la media 

```python
"""Esribe tu código aquí"""
open_prices = df['Open']
close_prices = df['Close']
average_price = (open_prices + close_prices) / 2
plt.figure(figsize=(10, 6))
plt.plot(df.index, close_prices, label='Open')
plt.fill_between(df.index, open_prices, close_prices, where=(close_prices > open_prices), color='blue', alpha=0.3, interpolate=True)
plt.fill_between(df.index, open_prices, close_prices, where=(close_prices <= open_prices), color='blue', alpha=0.3, interpolate=True)
plt.axhline(average_price.mean(), color='green', linestyle='--', label='Average Price')

```

```python
#Si creamos un nuevo df y metemos los vaores medios de TEL
df['TEL']=(df['Close']+df['Open'])/2
df_ibex35=pd.DataFrame(df['TEL'])
df_ibex35.head()
```

```python
#ahora metemos los de AENA
df =  yf.download('CABK',start,end)
df.head()
df['CABK']=(df['Close']+df['Open'])/2
df_ibex35['CABK']=df['CABK']
df_ibex35.head()
```

```python
#Haz que se descarque los valores de la siguiente lista 
lista=['IAG', 'MT', 'AMS', 'BBVA', 'BKT', 'TEF', 'SAN', 'ENG']
listaBuena=['MT', 'AMS', 'BBVA', 'BKT', 'TEF', 'SAN', 'ENG']
#Haz que los guarde dentro de una dataframe llamado lista_ibex35
lista_ibex35 = pd.DataFrame()
lista_ibex35 = yf.download('IAG', start=start, end=end)
lista_ibex35.head
for symbol in listaBuena:
    df = yf.download(symbol, start=start, end=end)
    lista_ibex35 = lista_ibex35 + df
lista_ibex35.head()
```

```python
#Visualizalo descomentando la siguiente instrucción
lista_ibex35.plot()
plt.figure(figsize=(10, 6))

for symbol in listaBuena:
    df = yf.download(symbol, start=start, end=end)
    plt.plot(df.index, df['Close'], label=symbol)
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Close Prices Over Time for Selected Symbols')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
